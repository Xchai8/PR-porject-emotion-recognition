```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import pickle
import joblib

from evaluate import Evaluate_performance
from tensorflow import keras
```

```python
test_data = np.load('Available data/test_data.npy')
test_label = np.load('Available data/test_label.npy')

test_data.shape, test_label.shape
```

```python
## knn
knn = joblib.load('trained_model/knn.pkl')

Evaluate_performance(model_type='knn', model=knn, history=None, test_data=test_data, test_label=test_label, display=False)
```

```python
## knn with PCA
knn_pca = joblib.load('trained_model/knn_pca.pkl')

Evaluate_performance(model_type='knn', model=knn_pca, history=None, test_data=test_data, test_label=test_label, display=False)
```

```python
ann = keras.models.load_model('trained_model/ann.h5')
with open('trained_model/ann_history.txt','rb') as file_pi:
    ann_history = pickle.load(file_pi)

Evaluate_performance(model_type='nn', model=ann, history=ann_history, test_data=test_data, test_label=test_label, display=True)
```

```python
cnn = keras.models.load_model('trained_model/cnn.h5')
with open('trained_model/cnn_history.txt','rb') as file_pi:
    cnn_history = pickle.load(file_pi)

Evaluate_performance(model_type='nn', model=cnn, history=cnn_history, test_data=test_data, test_label=test_label, display=True)
```

```python
lstm = keras.models.load_model('trained_model/lstm.h5')
with open('trained_model/lstm_history.txt','rb') as file_pi:
    lstm_history = pickle.load(file_pi)

Evaluate_performance(model_type='nn', model=lstm, history=lstm_history, test_data=test_data, test_label=test_label, display=True)
```
